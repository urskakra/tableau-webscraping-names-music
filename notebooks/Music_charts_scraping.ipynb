```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import numpy as np
```

```python
# a list of all years of top charts that we need (1920-2019)

year = list(map(str, range(1920, 2020)))
year
```

```python
# inside each year, get the rank, artist, and song

def single_year (year):
    
    url = 'https://playback.fm/charts/top-100-songs/'+ year
    
    html = requests.get(url)
    soup = BeautifulSoup(html.content, "html.parser")

    rank_ = soup.find_all("tr", {"itemprop": "track"})[0]
    rank = [i.getText().strip() for i in rank_]
    
    artist_ = soup.find_all("a", {"class": "artist"})[0]
    artist = [i.getText().strip() for i in artist_]

    song_ = soup.find_all("a", {"itemprop": "name"})[0]
    song = [i.getText().strip() for i in song_]

    rank_one = {
        
        'Year': year,
        'Rank': rank[1],
        'Artist': artist,
        'Song': song
    }
    
    return pd.DataFrame(rank_one)
```

```python
single_year ('1920')
```

```python
def top_charts (years):
    
    all_years = pd.DataFrame()
    
    for year in years:
    
        df = single_year(year)
        
        all_years = pd.concat([all_years, df], ignore_index=True)
            
    return pd.DataFrame(all_years)
```

```python
top_charts = pd.DataFrame(top_charts(year))
top_charts
```

```python
# convert Year column to integer

top_charts['Year'] = top_charts['Year'].apply(lambda x: int(x))
```

```python
# add a column with decade

top_charts['Decade'] = top_charts['Year'].apply(lambda x: x//10*10)
top_charts
```

```python
#save df to csv

top_charts.to_csv('../data/top_charts.csv', index=False)
```
